Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a = int(input())
b = int(input())
c = int(input())
if a+b>c and a+c>b and b+c>a: 
    print("la tam giac")
elif a==b or  a==c or b==c: 
    print("tam giac can")
elif a==b==c: 
    print("tam giac deu") 
elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
    print("tam giac vuong")
else:
    print("khong phai la tam giac")
```

bài 8 giải hệ phương trình tuyến tính 
$$
\begin{cases} 
ax + by = m \\
cx + dy = n 
\end{cases}
$$

```python
a, b, m = float(input()), float(input()), float(input())
c, d, n = float(input()), float(input()), float(input())

# Tính định thức
D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c

if D != 0:
    print(f"He co nghiem duy nhat: x = {Dx/D}, y = {Dy/D}")
else:
    if Dx == 0 and Dy == 0:
        print("He co vo so nghiem")
    else:
        print("He vo nghiem")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?
