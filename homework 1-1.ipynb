bai 3 Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem
trước cấu trúc lựa chọn: if….else, ma)

```python
#include <stdio.h>

int main() {
    char c;
    scanf("%c", &c);

    if (c >= 'A' && c <= 'Z')
        printf("%c", c + 32);
    else if (c >= 'a' && c <= 'z')
        printf("%c", c - 32);
    else
        printf("%c", c);

    return 0;
}
```

bai 4 Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c}
không phải là kí tự alphabet. 


```python
#include <stdio.h>

int main() {
    char c;
    scanf("%c", &c);

    if ((c >= 'A' && c <= 'Z') || (c >= 'a' && c <= 'z'))
        printf("%c la ki tu alphabet", c);
    else
        printf("%c khong phai la ki tu alphabet", c);

    return 0;
}
```

bai 6 Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
import math

a = float(input())
b = float(input())
c = float(input())

if a + b > c and a + c > b + b > a:
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(round(s, 1))
else:
    print("Khong phai 3 canh cua tam giac")
```

bai 9 Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
n = int(input())
match n:
    case 0: print("Không")
    case 1: print("Một")
    case 2: print("Hai")
    case 3: print("Ba")
    case 4: print("Bốn")
    case 5: print("Năm")
    case 6: print("Sáu")
    case 7: print("Bảy")
    case 8: print("Tám")
    case 9: print("Chín")
    case _: print("Không hợp lệ")
```
