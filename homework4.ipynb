W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input())
if n <= 0:
    print("sai")
else:
    k = 1
    while n >= 2*k:
        k = k + 1
        if n % 2*k == 0:
            print("dung")
        else:
            print("sai")
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
so_chan = 0
so_le = 0 
while n > 0:
    so = n % 10
    if so % 2 == 0:
        so_chan = so_chan + 1
    else:
        so_le = so_le + 1
    n = n // 10
print("so chan la", so_chan)
print("so le la", so_le)
```

1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
A = int(input())
n = 0
tong = 0 
while tong <=  A:
    n = n + 1
    tong = tong + 1/n 
print(n)
```

W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
A = int(input())
if A > 0:
    f1 = 1
    f2 = 1
    fib_max = 1
    while f2 <= A:
        f1, f2 = f2, f1 + f2 
        fib_max = f1
print(fib_max)
```
