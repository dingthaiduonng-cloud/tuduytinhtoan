W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def so_nguyen_to(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True 
n = int(input())
print(so_nguyen_to(n))    
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python
def so_hoan_hao(n):
    tong = 0 
    for i in range(1,n):
        if n % i == 0:
            tong = tong + i
    if tong == n:
        return True 
    else:
        return False
n = int(input())
print(so_hoan_hao(n))
```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 


```python
def giai_thua(n):
    tong = 1 
    for i in range(1, n + 1):
        tong = tong * i 
    return tong
n = int(input())
print(giai_thua(n))        
```

W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
def tong_chu_so(n):
    s = str(n)
    tong = 0 
    for i in s:
        tong = tong + int(i) 
    return tong  
n = int(input())
print(tong_chu_so(n))
```

Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor

```python
def divide_with_remainder(dividend, divisor):
    q = dividend // divisor 
    r = dividend % divisor 
    return q, r 
dividend, divisor = map(int,input().split())
print(divide_with_remainder(dividend, divisor))
```

W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op

```python
def compute(a, b, op):
    a = int(a)
    b = int(b)
    if op == '+':
        return a + b
    if op == '-':
        return a - b
    if op == '*':
        return a * b
    if op == '/':
        if b != 0:
            return a / b
        else:
            return "Lỗi chia cho 0"
    else:
        return "Toán tử không hợp lệ"
a,b,op = input().split()
print(compute(a, b, op))
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c):
    if a**2 == b**2 + c**2 or b**2 == a**2 + c**2 or c**2 == a**2 + b**2:
        return True 
    return False
a,b,c = map(int,input().split())
print(is_right_triangle(a,b,c))
```
