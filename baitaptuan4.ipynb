W4A1
[SumOfNumBers]
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để tính tống các số nguyên từ 1 đến n. (n <= 1000) 


```python
n = int(input())
i = 0 
tong = 0 
while i <= n:
    i = i + 1 
    tong = tong + i 
print(tong)
```

[PrimeNumber]
VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. 

```python
while True:
    n = int(input())
    if n <= 0:
        print("nhap lai")
    if n > 0:
        break 
```

W4A3
[Factorial]
Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n) với 0 < n < 100.

```python
n = int(input())
i = 1 
tong = 1
while 0 < i < n  < 100:
    i = i + 1 
    tong = tong * i 
print(tong)
```

W4A4
.[CountingDigits]
Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu)
của số đó. Không dùng kiểu dữ liệu string.


```python
n = int(input())
n = abs(n)
tong = 0 
if n == 0:
    print("1")
while n > 0: 
    tong = tong + 1 
    n = n // 10
print(tong)
```
