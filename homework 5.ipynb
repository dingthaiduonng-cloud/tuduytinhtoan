Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x 

```python
def so_amstrong(x):
    chu_so_x = str(x)
    n = len(str(x))
    tong = 0
    for digit in chu_so_x:
        tong += (int(digit) ** n)
    return tong == x
x = int(input())
if so_amstrong(x):
    print("true")
else:
    print("false")
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên

```python
def so_doi_xung(n):
    if n < 0:
        return False
    so_ban_dau = n
    so_dao_nguoc = 0
    while n > 0:
        chu_so_cuoi = n % 10 
        so_dao_nguoc = so_dao_nguoc * 10 + chu_so_cuoi
        n = n // 10 
    return so_dao_nguoc == so_ban_dau
n = int(input())
if so_doi_xung(n):
    print("True")
else:
    print("False")
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.


```python
def quadratic_roots_count(a, b, c):
    delta = b**2 - 4*a*c 
    if delta > 0:
        return 2 
    if delta == 0:
        return 1
    else: 
        return 0 
a, b, c =  map(int,input().split())
print(quadratic_roots_count(a, b, c))
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    if abs(x1 - x2) * abs(y2 - y3) == abs(x2 - x3) ** abs(y1 - y2):
        return "thang hang"
    else:
        return "khong thang hang"
x1,y1,x2,y2,x3,y3 = map(int,input().split())
print(are_collinear(x1,y1,x2,y2,x3,y3))
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)


```python
def khac_khong(n):
    dem = 0
    for so in n:
        chuoi = str(so)
        for chu_so in chuoi:
            if chu_so != "0":
                dem += 1
    return dem
n = list(map(int, input().split()))
print(khac_khong(n))
```

W5A29 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].


```python
def gia_tri(a,b,c):
    gia_tri_max = max(a,b,c)
    gia_tri_min = min(a,b,c)
    return [gia_tri_max, gia_tri_min]
a,b,c = list(map(int,input().split()))
print(gia_tri(a,b,c))
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k).


```python
import math
def he_so_nhi_thuc_C(n,k):
    if k > n:
        return False
    tong = math.factorial(n)/(math.factorial(k) * math.factorial(n-k))
    return tong 
n,k = map(int,input().split())
print(he_so_nhi_thuc_C(n,k))
```
